\begin{gather}
\begin{pmatrix} 1 && 1 && 0 \\ 1 && 0 && 1 \\ 0 && 1 && 1
\end{pmatrix} \begin{pmatrix} x_1 \\ x_2 \\ x_3\end{pmatrix} = 
\begin{pmatrix} y_1 \\ y_2 \\ y_3 \end{pmatrix}
\end{gather}

\begin{gather}
y_1 = \hat{x_1} + \hat{x_2}\\
y_2 = \hat{x_1} + \hat{x_3}\\
y_3 = \hat{x_2} + \hat{x_3}\\
\hat{x_1} \sim \mathcal{P}(x_1)\\
\hat{x_2} \sim \mathcal{P}(x_2)\\
\hat{x_3} \sim \mathcal{P}(x_3)
\end{gather}


---
Example: How to get the $x_1$?  
$x_{1_{recons}} \approx \frac{y_1 + y_2 - y_3}{2}$; there is 1 addition and 1 subtraction.

\begin{gather}
2x_{1_{recons}} = \hat{x_1} + \hat{x_1} + (\hat{x_2} - \hat{x_2}) + (\hat{x_3} - \hat{x_3})  
\end{gather}
So the final reconstructed value is the sum of two Poisson distribution and two Skellam distribution. It is still a Skellam distribution.
\begin{gather}
2x_{1_{recons}} = (\hat{x_1} + \hat{x_1} + \hat{x_2} + \hat{x_3}) - (\hat{x_2} + \hat{x_3})\\
\mu = 2x_1\\
\sigma^2 = 2x_1 + 2x_2 + 2x_3 \\
{SNR} \approx \frac{x_1}{\sqrt{3x_1}} = \sqrt{x_1} / 1.7
\end{gather}




```
N_variables = 500;
raw = 10;
a = np.ones(N_variables) * raw;
N_operations = 100;
operations = [1] * N_operations + [-1] * N_operations;
np.random.shuffle(operations);

obeservation = np.random.poisson(a);
for i in range(len(operations)):
    if operations[i] == 1:
        obeservation += np.random.poisson(a);
    elif operations[i] == -1:
        obeservation -= np.random.poisson(a);

plt.figure();
plt.hist(obeservation, bins = int(N_variables/20), density = True);
plt.xlim(-1000,1000)
plt.show();
```
